```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits = pd.read_csv("//content//tmdb_5000_credits.csv")
movies = pd.read_csv("//content//tmdb_5000_movies.csv")
```

```python
# merging the dataframe on basis of title
movies = pd.merge(movies,credits,on="title")
```

```python
# keeping columns
# genres , id , keywords , title , overview , cast , crew (director)v-> basis of recommendation

#populairty can be taken , release_date , vote_average

movies = movies[["genres" , "movie_id" , "keywords" , "title" , "overview" ,"cast" , "crew"]]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies['genres'].unique
```

```python
 # cleaning genres column
import ast
def clean_genres(obj):
  genreees = []
  for i in ast.literal_eval(obj):
    genreees.append(i['name'])
  return genreees
```

```python
movies['genres'] = movies['genres'].apply(clean_genres)
```

```python
movies.head(1)
```

```python
#cleaning  keyword with help of cleaning genres function
movies['keywords'] = movies['keywords'].apply(clean_genres)
```

```python
movies.head(1)
```

```python
# genres and keywords string dict converted to list type
```

```python
movies['cast'][1]
```

```python
# cleaning cast -> chosing top 5 actors or actresses from the cast
#to get the names of the top 5 actors or actresses from the movie (cast column is in form of string hence,coverting to list type)
def clean_cast(obj):
  castt = []
  cnt=1
  for i in ast.literal_eval(obj):
    if cnt<=5:
      castt.append(i['name'])
      cnt+=1
    else:
      break
  return castt
```

```python
movies['cast'] = movies['cast'].apply(clean_cast)
```

```python
movies.head(1)
```

```python
movies['cast'][1]
```

```python
# now handling crew column
movies['crew'][0]
```

```python
# checking for those names whose job is director from the list
def crew_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break

  return l
```

```python
movies['crew']=movies['crew'].apply(crew_director)
```

```python
movies.head()
```

```python
#overview -> gives the description of movie
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now removing all the spaces in between a name or name of genres to treat them as single entity and remove any futre problems
# eg. 'sam worthington' to 'samworthington'
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_movies = movies[['movie_id','title','tag']]
```

```python
new_movies.head(5)
```

```python
# convert each word of list in tag to string
new_movies['tag']=new_movies['tag'].apply(lambda x:" ".join(x))  # joining each list with with space
```
